```python
import requests
import pandas as pd
import time

# --- CONFIGURATION ---
API_KEY = "17308bb00e7328ad1f771a38aae3ae62"
BASE_URL = "https://api.themoviedb.org/3/discover/movie"

PARAMS = {
    "api_key": API_KEY,
    "with_genres": "878",
    "original_language": "en",
    "region": "US",
    "sort_by": "popularity.desc",
    "page": 1
}

def fetch_all_scifi_movies():
    movies_data = []
    max_tmdb_pages = 500 
    
    print("Lancement de l'extraction (jusqu'à 10 000 films)...")
    
    for page in range(1, max_tmdb_pages + 1):
        PARAMS['page'] = page
        response = requests.get(BASE_URL, params=PARAMS)
        
        if response.status_code != 200:
            print(f"Limite atteinte ou erreur API à la page {page}.")
            break
            
        data = response.json()
        results = data.get('results', [])
        total_pages = data.get('total_pages', 0)
        
        if not results:
            break # Il n'y a plus aucun film à récupérer
            
        for movie in results:
            movies_data.append({
                "title": movie.get("title"),
                "release_date": movie.get("release_date"),
                "popularity": movie.get("popularity"),
                "vote_average": movie.get("vote_average"),
                "vote_count": movie.get("vote_count")
            })
            
        print(f"✅ Page {page}/{min(max_tmdb_pages, total_pages)} récupérée.")
        
        if page >= total_pages:
            print("🏁 Fin absolue du catalogue atteinte !")
            break
        time.sleep(0.25) 
        
    return movies_data

# --- EXECUTION ---
if __name__ == "__main__":
    scifi_movies = fetch_all_scifi_movies()
    
    df = pd.DataFrame(scifi_movies)
    df = df.dropna(subset=['release_date'])
    
    # On sauvegarde tout
    df.to_csv("hollywood_scifi_movies_INTEGRAL.csv", index=False)
    
    print(f"\nTerminé ! {len(df)} films ont été sauvegardés.")
```
